```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("./finbert_local")
model = AutoModelForSequenceClassification.from_pretrained("./finbert_local")
```

```python
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
texts = [
    "The ECB decided to raise interest rates.",
    "The economic outlook remains strong and stable.",
    "Inflation pressures are decreasing slowly."
]

for t in texts:
    print(t, "→", nlp(t))
```
