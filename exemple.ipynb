```python
# analyse_finbert.py

from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Étape 1 : nom du modèle FinBERT
model_name = "ProsusAI/finbert"

# Étape 2 : télécharger automatiquement le tokenizer et le modèle
print("Téléchargement et chargement du modèle FinBERT...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Étape 3 : créer un pipeline pour simplifier l’analyse
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Étape 4 : exemple de texte BCE
text = "The ECB decided to raise interest rates to curb inflation."
result = nlp(text)

# Étape 5 : afficher le résultat
print(result)
```

```python
# Sauvegarde locale
model.save_pretrained("./finbert_local")
tokenizer.save_pretrained("./finbert_local")

print("Modèle sauvegardé dans ./finbert_local")
```
